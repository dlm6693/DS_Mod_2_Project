```python
def weather():
    base_url = 'https://api.darksky.net/forecast/'
    key = 'cab4d327b1b00264becd3f89ea6eb57b'
    lat_long = '40.7128,-74.0060'
    year_month = '2018-10-'
    rest = 'T00:00:00?America/New_York&exclude=flags,currently,daily,alerts, minutely'
    urls = []
    for i in range(1, 32):
        if len(str(i)) == 1:
            url = base_url+key+'/'+lat_long+','+year_month+'0'+str(i)+rest
            urls.append(url)
        elif len(str(i)) == 2:
            url = base_url+key+'/'+lat_long+','+year_month+str(i)+rest
            urls.append(url)
    weather = []
    for url in urls:
        r = requests.get(url)
        data = r.json().get('hourly').get('data')
        weather.append(data)
    flat_weather = [x for sublist in weather for x in sublist]
    return flat_weather
weather = weather()
```
